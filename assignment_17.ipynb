```python
student = {'name': 'Tharak', 'age': 22, 'course': 'Java Full Stack'}
```

```python
# 1️ clear() – removes all items from dictionary


student.clear()
print("After clear():", student)
```

```python
# 2️copy() – returns a shallow copy of dictionary
new_student = student.copy()
print("After copy():", new_student)
```

```python
# 3️ fromkeys() – creates a new dictionary from given keys
keys = ['name', 'age', 'course']
new_dict = dict.fromkeys(keys, 'unknown')
print("fromkeys():", new_dict)
```

```python
# 4️ get() – returns value of specified key
print("get('name'):", student.get('name'))
print("get('grade', 'Not Found'):", student.get('grade', 'Not Found'))
```

```python
# 5️ items() – returns list of (key, value) pairs
print("items():", student.items())
```

```python
# 6️ keys() – returns list of keys
print("keys():", student.keys())
```

```python
student = {'name': 'Tharak', 'age': 22, 'course': 'Python'}

age = student.pop('age')
print("After pop('age'):", student)
print("Popped value:", age)

```

```python
# 8️ popitem() – removes last inserted item (Python 3.7+)
last = student.popitem()
print("After popitem():", student)
print("Popped item:", last)
```

```python
# 9️ setdefault() – returns value of key, inserts key with default value if not present
course = student.setdefault('course', 'Python')
print("setdefault('course'):", course)
print("After setdefault:", student)

```

```python
# 10 update() – updates dictionary with another dictionary or key-value pairs
student.update({'age': 22, 'city': 'Hyderabad'})
print("After update():", student)
```

```python
# 1️1 values() – returns all values
print("values():", student.values())
```
