```python
# Q1.Sum of Three Numbers
 
def sum_three(a, b, c):
    """Returns the sum of three numbers."""
    return a + b + c


```

```python
print("Q1: Sum of Three Numbers =", sum_three(3, 5, 7))
```

```python
#Q2 Area of a Circle

def area_circle(radius):
    """Returns the area of a circle given its radius."""
    pi = 3.14159
    area = pi * (radius ** 2)
    return area
```

```python
print("Q2: Area of Circle =", area_circle(7))
```

```python
# Q3️ Check Even or Odd

def check_even_odd(num):
    """Prints whether a number is Even or Odd."""
    if num % 2 == 0:
        print("Even")
    else:
        print("Odd")
```

```python
print("Q3: Check Even or Odd:")
check_even_odd(5)    # ➡ Odd
check_even_odd(8)    # ➡ Eve
```

```python
#Q4️ Calculate BMI

def calculate_bmi(weight, height):
    """
    Returns the Body Mass Index (BMI).
    Formula: BMI = weight / (height ** 2)
    """
    bmi = weight / (height ** 2)
    return round(bmi, 2)
```

```python
print("Q4: BMI =", calculate_bmi(70, 1.75))
```

```python
# Q5 Greet with Name and Age

def greet(name, age):
    """Prints a greeting message with the user's name and age."""
    print(f"Hello {name}, you are {age} years old!")
```

```python
print("Q5: Greeting:")
greet("Aarav", 20)   # ➡ Hello Aarav, you are 20 years old!
```

```python
# Find Maximum of Two Numbers
def find_max(a, b):
    return a if a > b else b
```

```python
print(find_max(4, 9))
```

```python
#Power Function
def power(base, exponent):
    return base ** exponent
```

```python
print(power(2, 3))
```

```python
# Q8 Calculate Simple Interest

def simple_interest(principal, rate, time):
    return (principal * rate * time) / 100
```

```python
print(simple_interest(1000, 5, 2))
```

```python
# Q9 Check Leap Year
def is_leap_year(year):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        return True
    else:
        return False
```

```python
print(is_leap_year(2020)) 
print(is_leap_year(2023))  
```
