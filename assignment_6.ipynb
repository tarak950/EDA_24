```python
# Check if a number is even or odd
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("The number is even.")
else:
    print("The number is odd.")
```

```python
# Check whether a number is positive, negative, or zero
num = float(input("Enter a number: "))

if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
```

```python
# Print the larger of two numbers
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))

if a > b:
    print("The larger number is:", a)
elif b > a:
    print("The larger number is:", b)
else:
    print("Both numbers are equal.")
```

```python
# Check voting eligibility

age = int(input("Enter your age: "))

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

```python
# Check if a number is divisible by both 5 and 11

num = int(input("Enter a number: "))

if num % 5 == 0 and num % 11 == 0:
    print("The number is divisible by both 5 and 11.")
else:
    print("The number is not divisible by both 5 and 11.")
```

```python
# Check if a number is divisible by both 3 and 5

num = int(input("Enter a number: "))

if num % 3 == 0 and num % 5 == 0:
    print("The number is divisible by both 3 and 5.")
else:
    print("The number is not divisible by both 3 and 5.")
```

```python
# ATM-like withdrawal check

balance = float(input("Enter your account balance: "))
amount = float(input("Enter withdrawal amount: "))

if amount > balance:
    print("Withdrawal failed: Insufficient balance.")
elif amount % 100 != 0:
    print("Withdrawal failed: Amount must be a multiple of 100.")
else:
    print("Withdrawal Successful.")
```
