```python
# two numbers sub,mult,div
n1=eval(input('enter n1:'))
n2=eval(input('enter n1:'))
print(f"the subtraction of {n1},and{n2}, is {n1-n2}")
print(f"the multiplication of {n1},and{n2}, is{n1*n2} ")
print(f"the division of {n1},and{n2}, is {n1/n2}")
```

```python
# area of circle
pi=eval(input('enter value of pi:'))
r=eval(input('enter value of r:'))
area_of_circle=pi*r*r
print("area of circle:",area_of_circle)
```

```python
#area of triangle
b=eval(input('enter value of b:'))
h=eval(input('emter value of h:'))
area_of_triangle=0.5*b*h
print("area of triangle is:",area_of_triangle)

```

```python
n1=eval(input('enter n1:'))
n2=eval(input('enter n2:'))
n3=eval(input('enter n3:'))
avg=(n1+n2+n3)/3
avg1=round(avg,2)
print(f'the average of {n1} {n2},and{n3},is:{avg1}')
```

```python
# bill+tip calcultor
bill=eval(input('enter bill amount:'))
tip=eval(input('enter tip amount:'))
total_bill=bill+tip
print("total bill:",total_bill)
```

```python
# salary and tax calculator
salary=eval(input('enter salary:'))
tax_percentage=eval(input('enter tax:'))
tax_amount=(tax_percentage/100)*salary
print("tax amount:",tax_amount)
```
